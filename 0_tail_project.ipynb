# Study on the Tail Distribution of Stock Returns

## 1. Research Motivation and Objectives

Financial returns are often modeled as normally distributed, yet empirical evidence suggests the presence of extreme outcomes (“fat tails”).  
This study aims to describe the empirical distribution of stock returns and investigate how extreme returns are distributed across stocks, time, and sectors.
Building on this statistical characterization, we then explore whether extreme returns can be partially explained or anticipated using momentum indicators and sentiment-based signals derived from textual data. Machine learning methods will be employed to evaluate the predictive power of these features.

### Core Research Questions

1. How do you describe fat-tails?
   1. What statistical measures (kurtosis, tail ratios, extreme quantiles) best capture deviations from normality?
2. Do stock returns in a broad market exhibit fat tails?
   1. How does the empirical distribution compare to the normal benchmark?
3. What are these returns and what characteristics do they have
   1. Are extreme returns persistent at the individual stock level over time?
   2. What types of stocks contribute most to extreme returns?
   3. Are certain stock characteristics associated with a higher probability of extreme outcomes?
4. What financial / economic mechanisms may explain the presence of fat tails?
   1. What are the explainable variables, and what significance do they have?
   2. Do technical indicators (e.g., momentum measures) increase the likelihood of extreme outcomes?
   3. Does market sentiment, such as headline classification as a significance?
5. What are the implications for diversification and portfolio construction? (Prescriptive but might be too far.)
6. Does this behavior happens because of human psychology or information flow and lags?
   1. Is it possible to find other model behaving the same way, or building one?

## Initial analysis

The S&P 500 index is a natural starting point for the empirical analysis of stock return distributions. It represents a broad, liquid, and well-documented segment of the U.S. equity market, with extensive historical price data and a large cross-section of stocks, making it particularly suitable for studying distributional properties such as skewness, kurtosis, and tail behavior. Its prominence in both academic research and practical portfolio management also facilitates comparison with existing literature.

For practical reasons, this study relies on the current list of S&P 500 constituents and analyzes their historical returns over the selected sample period. This fixed-universe approach introduces survivorship bias, as stocks that were removed from the index due to poor performance or other factors are excluded from the analysis. Consequently, the estimated return distributions may overstate average performance and understate downside tail risk. This limitation is acknowledged throughout the study, and all conclusions should be interpreted as conditional on the surviving large-cap universe rather than the full historical index composition. Furthermore, I want fifteen years of history and as such, stocks that were added after 2010 will be discarded.
