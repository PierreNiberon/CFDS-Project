```python
import pandas as pd

# Wikipedia URL for S&P 500 constituents
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Read all tables from the page
tables = pd.read_html(url, storage_options={'User-Agent': 'Mozilla/5.0'})

# The first table is the constituents table
sp500 = tables[0]

# Keep only relevant columns
sp500_clean = sp500[[
    "Symbol",
    "Security",
    "GICS Sector",
    "GICS Sub-Industry"
]].copy()

# Rename columns for clarity
sp500_clean.columns = [
    "ticker",
    "company_name",
    "sector",
    "industry"
]

# Add index identifier (useful for later merges)
sp500_clean["index"] = "S&P500"

# Fix ticker format for yfinance (BRK.B -> BRK-B, BF.B -> BF-B)
sp500_clean["ticker"] = sp500_clean["ticker"].str.replace(".", "-", regex=False)

# Manually add the S&P500 index

index_row = pd.DataFrame([{
    "ticker": "^GSPC",   # better for yfinance
    "company_name": "S&P500",
    "sector": "Index",
    "industry": "Index",
    "index": "S&P500"
}])

sp500_clean = pd.concat([sp500_clean, index_row], ignore_index=True)

print("Manually added S&P500 index.")

# Save to CSV
sp500_clean.to_csv("data/sp500_list.csv", index=False)

print(f"Saved {len(sp500_clean)} S&P 500 constituents.")


```
