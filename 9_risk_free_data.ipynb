```python
import pandas as pd
import numpy as np

df_rf = pd.read_csv("data/DTB3.csv")

# rename for clarity
df_rf = df_rf.rename(columns={
    "observation_date": "date",
    "DTB3": "yield"
})

# convert date
df_rf['date'] = pd.to_datetime(df_rf['date'])

```

```python
display(df_rf)
```

```python
# forward fill to not have NaN

df_rf['yield'] = pd.to_numeric(df_rf['yield'], errors='coerce')
df_rf['yield'] = df_rf['yield'].ffill()
```

```python
display(df_rf)
```

```python
# check the number of recorde per year for 3M t-bills

counts = df_rf.groupby(df_rf['date'].dt.year).size()
print(counts)
```
