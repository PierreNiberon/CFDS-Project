```python
from time import sleep

# get all the tickers from the list and check the corporate actions

df = pd.read_csv("data/sp500_list.csv")

tickers = df["ticker"].unique()

actions_list = []

for ticker in tickers:
    sleep(1) # sleeps for 1 sec to avoid overloading yfinance api
    stock = yf.Ticker(ticker)
    actions = stock.actions  # contains Dividends + Stock Splits
    
    if not actions.empty:
        actions["Ticker"] = ticker
        actions_list.append(actions.reset_index())

corporate_actions = pd.concat(actions_list, ignore_index=True)

# deal with date format
corporate_actions = corporate_actions.reset_index()
corporate_actions["Date"] = corporate_actions["Date"].dt.date

corporate_actions.to_csv("data/sp500_corporate_actions.csv", index=False)
```
